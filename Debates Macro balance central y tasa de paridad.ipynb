5) b) Analisis Monetarista del TC Pre Paso, Hoy y Fin de Año 2020. A que valores debiera estar el TC.
Para este punto hicimos MV = PY, donde el TC y Base Monetaria deberían estar perfectamente interrelacionados, afectandosé de manera inmediata y sin latencia. 
Para generar tal perspectiva podemos tomar dos caminos. Primero, que el spread porcentual entre Base Monetaria y Cotización Oficial este internalizado, eliminarlo. U el otro camino sería generar una valor fundamental de la moneda intrinseco, partiendo del balance del Banco Central sumamos Base Monetaria, Pasivos (Leliqs + Pases) y lo dividimos por el Stock de Reservas.

Veamos con que nos encontramos

```python
# Buscamos el spread entre las variables 
pct = pd.DataFrame(index=analisis.index)
pct['BaseMonetaria'] = analisis.BaseMonetaria.pct_change().cumsum()

pct['Cotizacion'] = reporte.TCRes.pct_change(1).cumsum()

spreadBMyTC = pct['BaseMonetaria'] - pct['Cotizacion'] # obtenemos el spread entre Base Monetaria y TC Oficial, que no internaliza el Tipo de Cambio

spread = spreadBMyTC.plot(figsize=(10,6))
spread.grid()
spreadBMyTC
```

```python
# Graficamos Base Monetaria y TC sin internalizar
TC = pct.fillna(method='ffill').plot(figsize=(10,6))
TC.grid()
```

Observamos que por fines 2018 a hoy el crecimiento porcentual del Tipo de Cambio ha sido superior al crecimiento de la Base Monetaria, siendo revertido tal escenario en fines de Marzo 2020.
La cotización monetarista sería de la siguiente manera

```python
# Si el Spread Base Monetaria fuera internalizado a la Cotización, este sería el TC Oficial

Monetarista = pd.DataFrame(index=reporte.index)
pct['SpreadBMyTC'] = pct['Cotizacion'] - pct['BaseMonetaria']
Monetarista['Vision'] = reporte['TCRes'] * (1 - pct['SpreadBMyTC'])
Mon = Monetarista.plot(figsize=(15,7))
Mon.grid()
```

```python
# Si queremos observarlo en variación porcentual sería el siguiente gráfico
Monetarista['BaseMonetaria'] = analisis['BaseMonetaria']
Mon = Monetarista.pct_change().cumsum().plot(figsize=(15,7))
Mon.grid()
```

```python
# Analisis Fundamental del TC Oficial. Partiendo del Balance sumamos Base Monetaria y Pasivos Total, dividido stock reservas

FXFundamental = (reporte.Pases + reporte.Leliqs + reporte.TotalBM) / reporte.StockReservas 
FXPesos = FXFundamental.plot(figsize=(10,6))
FXPesos.grid()
```

```python
# Sumemos los TCs con el Oficial para poder graficarlos
trescambios = pd.DataFrame(index=reporte.index)
trescambios['Oficial'] = reporte['TCRes']
trescambios['Fundamental'] = FXFundamental.values
trescambios['Monetarista'] = Monetarista['Vision'].values
tres = trescambios.plot(figsize=(15,7))
tres.grid()
```

```python
tresporcentual = trescambios.pct_change(1).cumsum().plot(figsize=(15,7))
tresporcentual.grid()
```

RESPUESTA: 
Por la visión Monetarista, observamos que si de haber internalizado el spread, el TC Oficial Pre PASO estaba alto (casi 10%),  ahora estaría bajo (casi un 60%) y para fin de año seguiría agravandosé de no revertir la política actual.

Bajo la visión Fundamental del valor de la cotización, observamos que pre Paso estaba subvaluado el Peso (casi 20%), que hoy esta subvaluado con una brecha de casi el 80%, al ser supuestos estaticos la condición actual se sostiene a fin de año.
